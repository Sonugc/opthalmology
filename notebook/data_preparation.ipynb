```python
left_labels = 'Left-Diagnostic Keywords'
right_labels = 'Right-Diagnostic Keywords'

left_fundus_keywords = df[left_labels].tolist()
right_fundus_keywords = df[right_labels].tolist()

all_keywords = left_fundus_keywords + right_fundus_keywords
print(all_keywords)
print(len(all_keywords))
```

```python

excluded_keywords = ["lens dust", "optic disk photographically invisible", "low image quality", "image offset"]
```

```python
# Remove specified keywords from both left and right columns
df['Left-Diagnostic Keywords'] = df['Left-Diagnostic Keywords'].apply(lambda x: '，'.join([kw for kw in x.split('，') if kw not in excluded_keywords]))
df['Right-Diagnostic Keywords'] = df['Right-Diagnostic Keywords'].apply(lambda x: '，'.join([kw for kw in x.split('，') if kw not in excluded_keywords]))
```

```python
# Save the updated DataFrame to the original CSV file, overwriting the existing file
csv_file_path = '../data/data_updated.csv'
df.to_csv(csv_file_path, index=False)
cleaned_data=pd.read_csv(csv_file_path)
```

```python
print(f'Updated CSV file created at: {csv_file_path}')
```

```python
BASE_DIR=r"data"
cleaned_data1 = cleaned_data[['ID', 'Left-Fundus','Right-Fundus', 'Left-Diagnostic Keywords', 'Right-Diagnostic Keywords','N','D','G','C','A','H','M','O']].copy()
csv_file_path = '../data/cleaned_data1.csv'
cleaned_data1.to_csv(csv_file_path, index=False)

# Function to randomly select one keyword from a string of keywords
def select_random_keyword(keyword_string):
    if isinstance(keyword_string, str):  # Check if the value is a string
        keywords = keyword_string.split('，')  # Split keywords by '，' delimiter
        return random.choice(keywords) if keywords else None  # Return a random keyword or None if no keywords
    else:
        return None  # Return None for non-string values (e.g., float, NaN)
    
# Apply the function to create new columns with randomly selected keywords
cleaned_data1['Left-Label'] = cleaned_data1['Left-Diagnostic Keywords'].apply(select_random_keyword)
cleaned_data1['Right-Label'] = cleaned_data1['Right-Diagnostic Keywords'].apply(select_random_keyword)
cleaned_data1.to_csv(csv_file_path, index=False)
```

```python
# create a mapping from keywords to class labels
class_labels = ['N','D','G','C','A','H','M','O']
keyword_label_mapping  = {
    'normal fundus':'N',
    'retinopathy':'D',
    'glaucoma':'G',
    'cataract':'C',
    'macular degeneration':'A',
    'hypertensive':'H',
    'myopia':'M',
    
}

# if the keyword contains label outside of the above then, label them as others 'O'
```

```python

def generate_label(diag_label):

     for label in keyword_label_mapping.keys():
         if label in diag_label :
              return keyword_label_mapping[label]
         else:
            return 'O'
```
